```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

from mood.dataset import MOOD_DATASETS, dataset_iterator, MOOD_REGR_DATASETS
```

```python
df = pd.DataFrame()
for name, (smiles, y) in dataset_iterator():
    df_ = pd.DataFrame({"Dataset": name, "Size": len(smiles), "Target": "Regression" if name in MOOD_REGR_DATASETS else "Classification"}, index=[0])
    df = pd.concat((df, df_), ignore_index=True)
df
```

```python
sns.set_theme(style="whitegrid")

df.sort_values("Size", inplace=True)

fig, ax = plt.subplots(figsize=(12, 6))
sns.barplot(data=df, x="Dataset", y="Size", hue="Target", ax=ax, dodge=False)
for tick in ax.get_xticklabels():
    tick.set_rotation(45)

ax.set_yscale("log")
ax.set_xlabel("")
ax.set_ylim(0, 50000)
ax.set_ylabel("Log Size")
plt.tight_layout()
```
